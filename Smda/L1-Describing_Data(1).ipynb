## The choice of the bin size is crucial to highlight/washout structure in the data!
#### Choice of binning can be motivated based on experimental characteristics (measurement resolution)
#### General rules: Same bin size (exceptions exists), enough counts per bin (typically more than 10)
## There is no best number of bins and different bin sizes can reveal different features of the data
#### Too little bins you loose information of your data 
#### Too many bins you get a comb patter, which also leads to loss of data information

## Some formulas

### Square root choice: $nbins=\sqrt{N}$
### Sturges formula: $nbins=1+\log_2N$
### Rice rule: $nbins=2\sqrt[3]{N}$
### Many more... 


## Some formulas

### Range = maximum value  minimum value

### Width of intervals =  Range / (# of intervals)

These formulas can then be used to create the frequency table followed by the histogram.
https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.pyplot.hist.html

```python
n=len(students) 
print("Student length is ", n)
myrang=max(students)-min(students)
print("Range is ", myrang)
number_intervals=np.sqrt(n)
width=myrang/number_intervals
print("My width is ", width)
binning=[]
for i in range(int(number_intervals)):
  binning.append(i*10)
  print(i,binning[i])
```

```python
plt.xlabel("age")
plt.ylabel("counts")
plt.title('My second histogram')
 
plt.hist(students, bins =binning)
plt.show()
```

# Mean & Variance

# Arithmetic mean is the most common mean and it is widely used to describe properties of data.

$\bar{x}=\frac{1}{N}\sum_i x_i$

### Other means exists (discussed in notes), like Geometric or harmonic means, but these are rarely used in physics. 

```python
# Example: Scott took 7 math tests with results
#Mean does not always belong to the array.
x=np.array([89,73,84,91,87,77,94])
print("mean value= ", x.mean())
```

### The mean gives us some information but it is not the end!
### How the data disperse around the mean?


### Lets first look at the average deviation

$$V(x)=\frac{1}{N} \sum_{i=1}^N (\bar{x}-x_i)$$

### Does it provide any usefull information?

### Another important quantity to describe the data is the variance

$$V(x)=\frac{1}{N} \sum_{i=1}^N (\bar{x}-x_i)^2$$

### it tells how scattered the data are around the mean
#### Alternative formula derived from the above:
$$V(x)=\frac{1}{N} \sum_{i=1}^N x_i^2-\left(\frac{1}{N} \sum_{i=1}^N x_i\right)^2=\overline{x^2}-\overline{x}^2$$

### One can also characterise the spread of the data around the mean using:
#### The range of the data (min and max values)
Outliers can be misleading and unreliable    
#### instead of square of the difference use the absolute value (Mean Absolute deviation)
Rarely used -- Hard to deal with derivatives of absolute values
#### Full Widths Half Max values (FWHM)
FWHM is independent of the tails (variance is very dependent)
For a Guassian distribution $FWHM=2.35\sqrt{V(x)}$

```python
from statistics import mean 

set1=np.array([3,4,4,5,6,8])
set2=np.array([1,2,4,5,7,11 ])

print("mean value set 1= ", set1.mean())
print("mean value set 2= ", set2.mean())
```

```python
print("Variance of sample set1 =", set1.var()) 
print("Variance of sample set2 =", set2.var())  
```

### Instead of the variance it is common to speak about root mean square (RMS) deviation $\sigma=\sqrt{V(x)}$. 
### This is also refered to as the standard deviation
This is widely used and has the same units and dimensions with $x$. 
$\sigma$ represents a reasonable amount for a particular datapoint to differ from the mean.

```python
print("RMS  of sample set1 =", np.sqrt(set1.var()))
print("RMS  of sample set2 =", np.sqrt(set2.var()))
```

## Other quantities typically used to describe the data are
### Mode: the most common value of a data set
### Median: the value that splits in 2 equal part the data set
### Higher powers of $x$:
#### Skew: $\gamma=\frac{1}{N\sigma^3}\sum_i(x_i-\overline{x})^3$ describing the tails
#### Kurtosis: $c=\frac{1}{\sigma^4}\overline{(x-\overline{x})^4}-3$ describing the peak relative to a Gaussian



```python
from statistics import median 

print("Median of data-set 1 is % s" % (median(set1))) 
print("Median of data-set 2 is % s" % (median(set2))) 
```

## Correlations

#### Given tow data sets X and Y, we can check if there are correlations between them.
#### We define the covariance

$cov(X,Y)=\frac{1}{N}\sum_i (X_i-\bar{X})(Y_i-\bar{Y})=\bar{XY}-\bar{X}\bar{Y}$

#### We also define a correlation coefficient (Pearson)
$r=\frac{cov(X,Y)}{\sigma_X\sigma_Y}$

```python
X=[1,3,2,5,8,7,12,2,4]
Y=[8,6,9,4,3,3,2,7,7]
```

```python
from numpy import cov
from statistics import variance
covariance = cov(X, Y)

print('Variance of X', variance(X))
print('Variance of Y', variance(Y))

print()
print('Covariance matrix')
print(covariance)
```

```python
from scipy.stats import pearsonr

print(pearsonr(X, Y)[0])
```
