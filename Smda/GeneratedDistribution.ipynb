```python
import random as rnd
import matplotlib.pyplot as plt
import numpy as np
import math

def myFunc(x):
    return 1.0+math.sin(2.0*x)

mydata=[]
i = 1
while i < 10000:
    valuex=rnd.uniform(0,2*math.pi)
    valuey=rnd.uniform(0,3.0)
    if valuey<myFunc(valuex):
        mydata.append(valuex)
        i += 1
   
y=[]
bins=int(math.sqrt(len(mydata))) 
myFuncPlot = np.arange(0.0, 2*math.pi, 0.02)
for i in myFuncPlot:
    y.append(len(mydata)/bins*myFunc(i))
    
plt.hist(mydata, bins)    
plt.xlabel('x')
plt.ylabel('counts')
plt.plot(myFuncPlot, y,'-.')
plt.show()
```
